combining two dataframes into a single excel file with seperate sheets

```python
april_month = {'Date' :['16/04/24','17/04/24','18/04/24','19/04/24'], #giving data by using dictionary
           'Temp' :[34,45,23,45],
           'Windspeed':[6,7,8,4],
           'weather':['rainy','good','snow','rainy']
               }
feb_month = [
    ('1/02/24',32,5,'rain'),
    ('2/02/24',34,9,'snow'),
    ('3/02/24',31,8,'snow'),
    ('4/02/24',29,6,'rain')
           ]
```

```python
# Convert the april_month dictionary to a Pandas DataFrame
april_month_df = pd.DataFrame(april_month)
feb_month_df = pd.DataFrame(feb_month)

# Write the DataFrames to the Excel file
with pd.ExcelWriter("Weather.xlsx") as writer:
    april_month_df.to_excel(writer, sheet_name='april')
    feb_month_df.to_excel(writer, sheet_name='Feburary')
```
