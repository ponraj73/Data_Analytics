```python
import json
import pandas as pd
from tabulate import tabulate

json_data = [
    {
        "name": "Raj",
        "age": 21,
        "marks": [
            {"english": 91},
            {"tamil": 99},
            {"math": 87},
            {"science": 89}
        ],
        "address": [
            {"20220101": "331 main road, chennai 600021"},
            {"20210101": "33 gandhi road, bangalore 600021"}
        ],
        "city": "chennai",
        "gender": "male",
        "salary": 200000
    },
    {
        "name": "Vim",
        "age": 28,
        "marks": [
            {"english": 93},
            {"tamil": 98},
            {"math": 99},
            {"science": 100}
        ],
        "address": [
            {"20190101": "11 MG road, chennai 600021"},
            {"20200101": "921 gandhi road, bangalore 600021"}
        ],
        "city": "bangalore",
        "gender": "female",
        "salary": 300000
    }
]

rows = []
for student in json_data:
    name = student["name"]
    age = student["age"]
    marks = {k: v for mark in student["marks"] for k, v in mark.items()}
    max_mark = max(marks.values())
    min_mark = min(marks.values())
    avg_mark = sum(marks.values()) / len(marks)
    recent_address = list(student["address"][-1].values())[0]
    city = student["city"]
    gender = student["gender"]
    salary = student["salary"]

    rows.append({
        "name": name,
        "age": age,
        "english": marks.get("english", None),
        "tamil": marks.get("tamil", None),
        "math": marks.get("math", None),
        "science": marks.get("science", None),
        "max_mark": max_mark,
        "min_mark": min_mark,
        "avg_mark": avg_mark,
        "recent_address": recent_address,
        "city": city,
        "gender": gender,
        "salary": salary
    })

df = pd.DataFrame(rows)
display(pd.DataFrame(df))

df.to_csv("student.csv", index=False)

print(tabulate(df, headers="keys", tablefmt="grid"))
```
