1)**Top 10 Type by # events. Type column is the event.**


```python
top_types = data['type'].value_counts().reset_index()
top_types.columns = ['type', 'num_events']
top_types['rank'] = top_types['num_events'].rank(method='min', ascending=False).astype(int)
top_types = top_types.head(10)
print("Top 10 Types by # Events")
print(tabulate(top_types, headers='keys', tablefmt='grid'))
```

2)**Top 10 Ethnics by # events. Use column “Officer-defined ethnicity"**

```python
top_officer_ethnicities = data['Officer-defined ethnicity'].value_counts().reset_index()
top_officer_ethnicities.columns = ['type', 'office_num_events']
top_officer_ethnicities['office_ethnic_rank'] = top_officer_ethnicities['office_num_events'].rank(
    method='min', ascending=False
).astype(int)
top_officer_ethnicities = top_officer_ethnicities.head(10)
print("\nTop 10 Ethnicities by # Events (Officer-defined ethnicity)")
print(tabulate(top_officer_ethnicities, headers='keys', tablefmt='grid'))
```

3)**Top 10 Ethnics by # events. Use column “Self-defined ethnicity”**

```python
top_self_ethnicities = data['Self-defined ethnicity'].value_counts().reset_index()
top_self_ethnicities.columns = ['type', 'self_num_events']
top_self_ethnicities['self_ethnic_rank'] = top_self_ethnicities['self_num_events'].rank(
    method='min', ascending=False
).astype(int)
top_self_ethnicities = top_self_ethnicities.head(10)
print("\nTop 10 Ethnicities by # Events (Self-defined ethnicity)")
print(tabulate(top_self_ethnicities, headers='keys', tablefmt='grid'))
```

4)**Merge 2 and 3 above on type.**

```python
ethnicities_merged = pd.merge(top_officer_ethnicities, top_self_ethnicities, on='type', how='inner')
print("\nMerged Ethnicity Data")
print(tabulate(ethnicities_merged, headers='keys', tablefmt='grid'))
```

5)**Top 10 Type by Gender = Male.**

```python
male_events = data[data['Gender'] == 'Male']['type'].value_counts().reset_index()
male_events.columns = ['type', 'num_events_men']
male_events['rank_men'] = male_events['num_events_men'].rank(method='min', ascending=False).astype(int)
male_events = male_events.head(10)
print("\nTop 10 Types by Gender (Male)")
print(tabulate(male_events, headers='keys', tablefmt='grid'))
```

6)**Top 10 Type by Gender = Female.**

```python
female_events = data[data['Gender'] == 'Female']['type'].value_counts().reset_index()
female_events.columns = ['type', 'num_events_women']
female_events['rank_women'] = female_events['num_events_women'].rank(method='min', ascending=False).astype(int)
female_events = female_events.head(10)
print("\nTop 10 Types by Gender (Female)")
print(tabulate(female_events, headers='keys', tablefmt='grid'))
```

7)**Group bar chart plot (vertical) with output from 5 and 6.**
[1][2]

**Bar 1 -> Male**

**Bar 2 -> Female**

```python
merged_gender_data = pd.merge(male_events, female_events, on='type', how='outer').fillna(0)
merged_gender_data.plot(x='type', kind='bar', y=['num_events_men', 'num_events_women'], figsize=(10, 6))
plt.title('Event Counts by Gender')
plt.ylabel('Number of Events')
plt.xlabel('Type')
plt.xticks(rotation=45)
plt.legend(['Male', 'Female'])
plt.show()
```

8)**For Gender = Male, # events by “Age range”**


```python
male_age_events = data[data['Gender'] == 'Male'].groupby(['type', 'Age range']).size().reset_index(name='num_events_male')
male_age_events['rank_male'] = male_age_events.groupby('type')['num_events_male'].rank(method='min', ascending=False).astype(int)
print("\nTop 10 Types by Gender (Male)")
print(tabulate(male_age_events, headers='keys', tablefmt='grid'))
```

9)**For Gender = Female, # events by “Age range”.**

*   List item
*   List item



```python
female_age_events = data[data['Gender'] == 'Female'].groupby(['type', 'Age range']).size().reset_index(name='num_events_female')
female_age_events['rank_female'] = female_age_events.groupby('type')['num_events_female'].rank(method='min', ascending=False).astype(int)
print("\nTop 10 Types by Gender (Female)")
print(tabulate(female_age_events, headers='keys', tablefmt='grid'))
```

10)**Merge 8 and 9 on column “type”.**

```python
age_events_merged = pd.merge(male_age_events, female_age_events, on='type', how='outer', suffixes=('_male', '_female')).fillna(0)
print("\nMerged Age Range Data for Male and Female")
print(tabulate(age_events_merged, headers='keys', tablefmt='grid'))
```

1) **The file "Employee_start_time_end_time.csv" contains the start and end times for employees' daily work shifts. The file "hour_range_slot_for_pay.csv" specifies the pay rate applicable based on the number of hours worked for a particular day. For example, if an employee works for 4 hours on a given day, the applicable pay rate is $10 per hour, as 4 hours falls within the range of 1 to 5 hours (see row 1 in the "hour_range_slot_for_pay.csv" file). If an employee works for 7 hours on a given day, the pay rate is $12 per hour, as it falls within the range of 6 to 10 hours (see row 2 in the "hour_range_slot_for_pay.csv" file). Using these criteria, you need to calculate the total pay in USD for each employee based on the provided information from the two CSV files.**

```python
import pandas as pd

df1 = pd.read_csv(r'/content/gdrive/My Drive/mini project/employee_start_time_end_time.csv')
df2 = pd.read_csv(r'/content/gdrive/My Drive/mini project/hour_range_slot_for_pay.csv')

combined_df = pd.concat([df1, df2], ignore_index=True)

combined_df.to_csv('combined_file.csv', index=False)
display(pd.DataFrame(combined_df))
df = pd.DataFrame(combined_df)
```

```python
import pandas as pd

# Load the employee work hours and pay rate CSV files
df_employee = pd.read_csv(r'/content/gdrive/My Drive/mini project/employee_start_time_end_time.csv')
df_pay_rate = pd.read_csv(r'/content/gdrive/My Drive/mini project/hour_range_slot_for_pay.csv')

# Convert start and end times to datetime
df_employee['start_time'] = pd.to_datetime(df_employee['start_time'])
df_employee['end_time'] = pd.to_datetime(df_employee['end_time'])

# Calculate the total hours worked for each employee
df_employee['hours_Worked'] = (df_employee['end_time'] - df_employee['start_time']).dt.total_seconds() / 3600

# Define a function to find the pay rate based on hours worked
def get_pay_rate(hours):
    for _, row in df_pay_rate.iterrows():
        if row['daily_hour_min_range'] <= hours <= row['daily_hour_max_range']:
            return row['pay_per_hour_usd']
    return 0  # Default rate if no range is matched

# Apply the function to calculate the pay rate for each employee
df_employee['pay_per_hour_usd'] = df_employee['hours_Worked'].apply(get_pay_rate)

# Calculate total pay
df_employee['Total_Pay'] = df_employee['hours_Worked'] * df_employee['pay_per_hour_usd']

# Save the results to a new CSV file
df_employee[['employee_id', 'Total_Pay']].to_csv(r'/content/gdrive/My Drive/mini project/employee_total_pay.csv', index=False)

# Display the result
display(df_employee[['employee_id', 'Total_Pay']])
```

2)**Credit Card Fraud Detection: Credit Card Fraud classification using Random Forest Classifier. Column value to be classified is in column "Is_fraud" and all other columns are the features. Use two-fold cross validation and evaluation metrics: Precision, Recall, F1-score, Accuracy [wiki]. Also print confusion matrix [wiki]. Refer basic notebook template for some code and also internet.**

```python
data= pd.read_excel(r'/content/gdrive/My Drive/mini project/credit_card_data.xlsx')
display(pd.DataFrame(data))
df = pd.DataFrame(data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix, classification_report

X = df.drop(columns=['Is_fraud'])
y = df['Is_fraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

rf_model = RandomForestClassifier(random_state=42)

skf = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)

precision_scores = []
recall_scores = []
f1_scores = []
accuracy_scores = []
confusion_matrices = []

for train_idx, val_idx in skf.split(X, y):

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    rf_model.fit(X_train, y_train)

    y_pred = rf_model.predict(X_val)

    precision_scores.append(precision_score(y_val, y_pred))
    recall_scores.append(recall_score(y_val, y_pred))
    f1_scores.append(f1_score(y_val, y_pred))
    accuracy_scores.append(accuracy_score(y_val, y_pred))
    confusion_matrices.append(confusion_matrix(y_val, y_pred))

print(f"Precision: {sum(precision_scores) / len(precision_scores):.4f}")
print(f"Recall: {sum(recall_scores) / len(recall_scores):.4f}")
print(f"F1-Score: {sum(f1_scores) / len(f1_scores):.4f}")
print(f"Accuracy: {sum(accuracy_scores) / len(accuracy_scores):.4f}")

print("\nConfusion Matrix (Last Fold):")
print(confusion_matrices[-1])

print("\nClassification Report (Last Fold):")
print(classification_report(y_val, y_pred))
```
