```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/Project files/minipro 2/us-counties-recent.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Project files/minipro 2/us-counties.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Project files/minipro 2/us-states.csv")
df4 = pd.read_csv("/content/drive/MyDrive/Project files/minipro 2/us.csv")
```

```python
df1.head(5)
df2.head(5)
```

```python
df_new = pd.merge(df1,df2, how="outer")
df_new.head(5)
```

(1) Top 10 states having highest total death in year 2020

```python
#Top 10 states having highest total death in year 2020

def getyear(row):
  return datetime.strptime(row["date"], "%Y-%m-%d").year

df_new["year"] = df_new.apply(getyear, axis=1)
df_in2020 = df_new.where(df_new["year"] == 2020)
df_top10 = df_in2020.groupby(["state"])["deaths"].sum().to_frame("count").reset_index()
df_top10_2020 = df_top10.sort_values(by="count", ascending=False)
df_top10_2020.head(10)
```

(2) Top 10 states having highest total death in the year 2021

```python
#Top 10 states having highest total death in the year 2021
df_new["year"] = df_new.apply(getyear, axis=1)
df_in2021 = df_new.where(df_new["year"] == 2021)
df_top10 = df_in2021.groupby(["state"])["deaths"].sum().to_frame("count").reset_index()
df_top10_2021 = df_top10.sort_values(by="count", ascending=False)
df_top10_2021.head(10)
```

(3) Top 10 states having highest avg death in the year 2020

```python
#Top 10 states having highest avg death in the year 2020
df_new["year"] = df_new.apply(getyear, axis=1)
df_in2020 = df_new.where(df_new["year"] == 2020)
df_top10 = df_in2020.groupby(["state"])["deaths"].mean().to_frame("count").reset_index()
df_top10_2020 = df_top10.sort_values(by="count", ascending=False)
df_top10_2020.head(10)
```

(4) Top 10 states having highest avg death in the year 2021

```python
#Top 10 states having highest avg death in the year 2021
df_new["year"] = df_new.apply(getyear, axis=1)
df_in2021 = df_new.where(df_new["year"] == 2021)
df_top10 = df_in2021.groupby(["state"])["deaths"].mean().to_frame("count").reset_index()
df_top10_2021 = df_top10.sort_values(by="count", ascending=False)
df_top10_2021.head(10)
```

(5) Top 10 county having highest total death in 2020 and 2021 year together

```python
#Top 10 county having highest total death in 2020 and 2021 year together
df_new["year"] = df_new.apply(getyear, axis=1)
dfboth2021_21 = df_new.where(df_new["year"] == 2020 & 2021)
df_top10 = dfboth2021_21.groupby(["state"])["deaths"].sum().to_frame("count").reset_index()
df_top10_2021_21 = df_top10.sort_values(by="count", ascending=False)
df_top10_2021.head(10)
```
