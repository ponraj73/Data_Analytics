(7) Print list of Top n states by the least number of death in a specific year

Top 10 least death states in 2020

```python
# (7) Print list of Top n states by the least number of death in a specific year
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year

df_new['year'] = df_new.apply(get_year, axis=1)
#filtering the data on 2020
df_2020 = df_new.where(df_new['year'] == 2020)
df_top10_2020 = df_2020.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
df_top10_2020 = df_top10_2020.sort_values(by=['count'], ascending=True)

def rank(row):
    return int( 56 - row['rank1'])

df_top10_2020['rank1'] = df_top10_2020['count'].rank(ascending = False )
df_top10_2020['rank'] = df_top10_2020.apply(rank, axis=1)
df_top10_2020_sel = df_top10_2020[['state', 'count', 'rank']]
df_top10_2020_sel.head(10)
```

Top 10 least death states in 2021

```python
def get_year(row):
    return datetime.strptime(row['date'], '%Y-%m-%d').year

df_new['year'] = df_new.apply(get_year, axis=1)
#filtering the data on 2021
df_2021 = df_new.where(df_new['year'] == 2021)
df_top10_2021 = df_2021.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
df_top10_2021 = df_top10_2021.sort_values(by=['count'], ascending=True)

def rank(row):
    return int( 56 - row['rank1'])

df_top10_2021['rank1'] = df_top10_2021['count'].rank(ascending = False )
df_top10_2021['rank'] = df_top10_2021.apply(rank, axis=1)
df_top10_2021_sel = df_top10_2021[['state', 'count', 'rank']]
df_top10_2021_sel.head(10)
```

(8) Top 10 least death states in last three months of the year 2021

```python
df_new['date'] = pd.to_datetime(df_new['date']) # Convert 'date' column to datetime objects
# Filter data for the last three months of 2021
df_last_3_months_2021 = df_new[(df_new['date'] >= '2021-10-01') & (df_new['date'] <= '2021-12-31')]
# Group by state and sum deaths for the last three months of 2021
df_top10_last_3_months_2021 = df_last_3_months_2021.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
# Sort by the total deaths in ascending order
df_top10_last_3_months_2021 = df_top10_last_3_months_2021.sort_values(by=['count'], ascending=True)
df_top10_last_3_months_2021['rank1'] = df_top10_last_3_months_2021['count'].rank(ascending=False) # Assign ranks

def rank(row):
    return int(56 - row['rank1'])

df_top10_last_3_months_2021['rank'] = df_top10_last_3_months_2021.apply(rank, axis=1)
# Select relevant columns
df_top10_last_3_months_2021_sel = df_top10_last_3_months_2021[['state', 'count', 'rank']]
# Display the top 10 states with the least deaths in the last three months of 2021
df_top10_last_3_months_2021_sel.head(10)
```

Top 10 most death states in last three months of the year 2021

```python
df_new['date'] = pd.to_datetime(df_new['date']) # Convert 'date' column to datetime objects
# Filter data for the last three months of 2021
df_last_3_months_2021 = df_new[(df_new['date'] >= '2021-10-01') & (df_new['date'] <= '2021-12-31')]
# Group by state and sum deaths for the last three months of 2021
df_top10_last_3_months_2021 = df_last_3_months_2021.groupby(['state'])['deaths'].sum().to_frame('count').reset_index()
# Sort by the total deaths in descending order
df_top10_last_3_months_2021 = df_top10_last_3_months_2021.sort_values(by=['count'], ascending=False)
# Assign ranks
df_top10_last_3_months_2021['rank1'] = df_top10_last_3_months_2021['count'].rank(ascending=False)

def rank(row):
    return int(row['rank1'])

df_top10_last_3_months_2021['rank'] = df_top10_last_3_months_2021.apply(rank, axis=1)
# Select relevant columns
df_top10_last_3_months_2021_sel = df_top10_last_3_months_2021[['state', 'count', 'rank']]
# Display the top 10 states with the most deaths in the last three months of 2021
df_top10_last_3_months_2021_sel.head(10)
```
